```python
import re

def clean_wiki(wiki_info):
    after_wiki = {}
    for key, value in wiki_info.items():
        after_wiki[key] = re.sub(r'(\[(\d+)+-*(\d+)*\]|\xa0+|（概述图+?(.*)?）)','',wiki_info[key])
        after_wiki[key] = re.sub(r'(拼音|读音|注音|读作)+?((是|：|为)*)?','',after_wiki[key])
        after_wiki[key] = re.sub(r'(释义|意思|泛指|亦命)+?((是(指)*|：|为)*)?','',after_wiki[key])
    return after_wiki
```

```python
wiki_info = clean_wiki(wiki_info)
wiki_info
```

```python
with open('data/cleaned_wiki.json', 'w') as f:
    json.dump(wiki_info, f, indent=2, ensure_ascii=False)
```
